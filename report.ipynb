Это сеть, решающая задачу улучшения речи (убирание шумов).

В этой статье больше привнесли нечто новое в GANы, нежели в саму задачу.
Просто SE (speech enhansement) оказалась одной из наиболее подходящих задач, чтобы продемонстрировать преимущества нового подхода.

В чем же основное нововведение?
Обычно при тренировке GAN лосс считаются так:

$$L_G = (D(G(x),x) - 1)^2$$
$$L_D = (D(G(x),x) - 0)^2 + (D(y,y) - 1)^2 $$

Авторы же статьи предлагают новый подход. Они утверждают, что если в задаче сложная метрика, то лучше учить генератор подстраиваться именно под метрику. 

Как это сделать? Очень просто. Теперь дискриминатор будет пытаться имитировать метрику, учась под нее, а генератор будет учиться под дискриминатор, то есть, по сути, под метрику.

Тогда новый лосс дискриминатора такой:

$$L_{Dnew} = (D(y,y) - 1)^2 + (D(G(x),y) - Q'(G(x),y))^2 $$

Где $Q' -$ метрика, причем $Q' \in [0,1]$

 Я успешно запустил GAN и обучал ~12ч. Сеть неплохо обрабатывает звуки, однако не без минусов - звук получается несколько "роботизированным" и неестественным. Метрика в среднем была 0.75, что, в принципе соответствует результатам статьи.
